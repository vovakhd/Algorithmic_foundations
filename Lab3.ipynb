```python
import numpy as np
import matplotlib.pyplot as plt

dataset = np.loadtxt("DS7.txt")
def affine_transform(point, angle, center):
    angle_rad = np.radians(angle)
    x, y = point
    cx, cy = center
    new_x = np.cos(angle_rad) * (x - cx) - np.sin(angle_rad) * (y - cy) + cx
    new_y = np.sin(angle_rad) * (x - cx) + np.cos(angle_rad) * (y - cy) + cy
    return new_y, new_x

center = (480, 480)
angle = 80
transformed_dataset = [affine_transform(point, angle, center) for point in dataset]
plt.figure(figsize = (960/80, 960/80))
plt.xticks([])
plt.yticks([])
plt.scatter(*zip(*transformed_dataset), color = 'blue')

plt.savefig("Lab3_image.png")
plt.show()
```
