```python
import matplotlib.pyplot as plt

def read_dataset(file_path):
    coordinates = []
    with open(file_path, 'r') as file:
        for line in file:
            x, y = map(int, line.strip().split())
            coordinates.append((y, x))
    return coordinates

def display_points(coordinates):
    x, y = zip(*coordinates)
    plt.scatter(x, y)
    plt.xlabel('X')
    plt.ylabel('Y')

def save_plot(file):
    plt.savefig(file)

def main():
    file_path = 'DS7.txt'
    canvas_size = (960, 540)

    coordinates = read_dataset(file_path)

    plt.figure(figsize = (canvas_size[0]/80, canvas_size[1]/80), dpi = 80)
    plt.xlim(0, canvas_size[0])
    plt.ylim(0, canvas_size[1])

    display_points(coordinates)
    save = 'Lab2_image.png'
    save_plot(save)

if __name__ == "__main__":
    main()
```
